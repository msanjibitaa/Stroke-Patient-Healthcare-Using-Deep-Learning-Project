Handling null values

```python
# Filling up the missing values with the value as "Unknown"
#df["bmi"] = df["bmi"].fillna("Unknown")
```

```python
# Filling up the missing values with the mean or median instead of "Unknown."
df['bmi'] = df['bmi'].fillna(df['bmi'].mean()) #or df['bmi].median()
```

Convert Categorical And Numerical Attributes to 'category' and 'int' Data Type

```python
#Convert Categorical Attributes:
categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
df[categorical_cols] = df[categorical_cols].astype('category')
                
```

```python
# List of numeric attributes to convert to integer type
num_attributes = ['hypertension', 'heart_disease', 'stroke']

# Convert specified columns to integer type
df[num_attributes] = df[num_attributes].astype(int)
```
